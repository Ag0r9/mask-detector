# Face detection testing

## Packages

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
test_face = cv2.imread('pictures/test_face.jpg')
```

```python
plt.imshow(convertToRGB(test_face))
```

```python
gray_img = cv2.cvtColor(test_face, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
haar_face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=2, minNeighbors=5)
```

```python
print('Faces found: ', len(faces))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(test_face, (x, y), (x+w, y+h), (0, 255, 255), 2)
```

```python
plt.imshow(convertToRGB(test_face))
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1):
    img_copy = colored_img.copy()
    gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)       
    faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5);   
    for (x, y, w, h) in faces:
        cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 255), 2)
    print
    return img_copy
```

```python
plt.figure(figsize=(20,10))
multi_faces = cv2.imread('pictures/melanrz.jpg')
faces_detected_img = detect_faces(haar_face_cascade, multi_faces)
plt.imshow(convertToRGB(faces_detected_img))
```
